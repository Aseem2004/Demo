**This is a case of Multiple Linear Regression where there would be more than one coefficients of the equation**

```python
import pandas as pd
```

```python
HousePrice=pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/Boston.csv")
```

 Here MEDV ie median house price is dependent on other 13 factors. Equation will be of the form **y=c+m<sub>1</sub>x<sub>1</sub>+m<sub>2</sub>x<sub>2</sub>+...+m<sub>13</sub>x<sub>13</sub>**.

---
There will be 13 coefficients/slopes here and a single intercept.


```python
y=HousePrice['MEDV']
x=HousePrice[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

These are the 13 variables for the equation of line.

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
 mean_absolute_percentage_error(y_test,y_pred)
```

It means approx 16.5% error and 83.5% accuracy of the model.
